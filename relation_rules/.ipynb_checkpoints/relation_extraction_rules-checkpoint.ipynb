```python
#!/user/bin/env python
# coding:utf-8
import re

#Read File and Genrate Corresponding Keywords
def gen_obv_keywords(infile_name):
    infile = open(infile_name)
    obv_rule_gen = set()
    for line in infile:
        line = line.strip().strip('"').strip()
        #print line
        if len(line) > 0:
            obv_rule_gen.add(line)
    infile.close()
    return obv_rule_gen

#Genrate Re String List
def gen_re_string(sub, obj, obv_rules):
    #rule_key: recognizer for different rules
    #obv_rules: a dictionary of rules, each has set of obviouswords for the current rule
    re_strings = []
    
    #Unicode Characters
    sub = unicode(sub, 'utf-8')
    obj = unicode(obj, 'utf-8')
    
    for rule_key in obv_rules:
        obv_string = u"["
        for obv in obv_rules[rule_key]:
            obv = unicode(obv, 'utf-8')
            obv_string = obv_string + obv + u","
        obv_string = obv_string.strip(u",") + u"]"
    
        #Add Rules
    
    
        #Rule Testing
        if rule_key == 'test':
            re_string = sub + u"[\u4e00-\u9fa5]{0,3}" + obv_string + obj
            re_strings.append(re_string)
    
        #Rule0
        if rule_key == 'pre_word':
            re_string = sub + u"[\u4e00-\u9fa5]{0,6}" + obv_string + obj
            re_strings.append(re_string)
            re_string = obj + u"[\u4e00-\u9fa5]{0,6}" + obv_string + sub
            re_strings.append(re_string)
    
        #Rule1
        if rule_key == 'action_word':
            re_string = sub + u"[\u4e00-\u9fa5]{0,6}" + obj + u"[\u4e00-\u9fa5]{0,3}" + obv_string
            re_strings.append(re_string)
            re_string = obj + u"[\u4e00-\u9fa5]{0,6}" + sub + u"[\u4e00-\u9fa5]{0,3}" + obv_string
            re_strings.append(re_string)
    
    return set(re_strings)

#Match Rules
def match_rule(target_string, sub, obj, obv_rules):
    #obv_rules: a dictionary of obvs, each of with is a set of obviouswords
    target_string = unicode(target_string, 'utf-8')
    re_strings = set()
    
    #Generate re_strings
    re_strings = gen_re_string(sub, obj, obv_rules)
    #print unicode(re_strings)
    #For each rule, match the target string
    for re_rule in re_strings:
        if re.match(re_rule,target_string) != None:
            #print sub
            #print obj
            #print target_string
            #print '#########################'
            return 1
        
    #print "Not Found"
    return 0

```
